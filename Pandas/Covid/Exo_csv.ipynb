```python
## Plusieurs groupby
data[['continent', 'location','new_cases', 'new_deaths']].groupby(['continent','location']).sum()
```

```python
## Largest new_cases in a month / location
data[['location','new_cases']][(data['date'] > '2020-06-24') & (data['date'] < '2020-07-24')].groupby(['location']).max()['new_cases'].nlargest()
```

```python
## Largest new_deaths in a month / location
data[['location','new_deaths']][(data['date'] > '2020-06-24') & (data['date'] < '2020-07-24')].groupby(['location']).max()['new_deaths'].nlargest()
```

```python
## Largest Ratio in new cases and deaths (population_density)
density=data[['location','new_cases','new_deaths','population_density']].groupby(['location','population_density']).sum()
```

```python
density = density.reset_index() # Permet de remettre les index Ã  0 ... (possible Ã  relancer)
```
